```python
import pandas as pd
import datetime
import numpy as np
import pandas.api.types as pdtypes


df = pd.read_csv("agosto23_bike.csv")

año_actual = datetime.datetime.now().year

df['Edad'] = año_actual - df['Anyo_de_nacimiento']

df = df.loc[df['Edad'] >= 18]

df['Edad'] = df['Edad'].astype(int)

df['Genero'] = df['Genero'].astype(pdtypes.CategoricalDtype(categories=['F', 'M']))

```
