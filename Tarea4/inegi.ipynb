```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('inegi.csv.CSV')

df
```

```python
print(df['estado'].unique())
```

```python
# Crear un histograma de la columna 'estado'
plt.figure(figsize=(15, 10))
df['estado'].value_counts().sort_index().plot(kind='bar', edgecolor='k', alpha=0.7)
plt.title('Distribución de Estados')
plt.xlabel('Estado')
plt.ylabel('Frecuencia')
plt.xticks(range(0, 32), ['AGS', 'BC', 'BCS', 'CAM', 'CHP', 'CHI', 'CDM', 'COA', 'COL', 'DUR', 'GTO', 'GRO', 'HGO', 'JAL', 'MEX', 'MIC', 'MOR', 'NAY', 'NLE', 'OAX', 'PUE', 'QRO', 'ROO', 'SLP', 'SIN', 'SON', 'TAB', 'TAM', 'TLX', 'VER', 'YUC', 'ZAC' ])
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Crear un histograma de barras para la columna 'fn_mes'
plt.figure(figsize=(10, 6))
df['fn_mes'].value_counts().sort_index().plot(kind='bar', edgecolor='k', alpha=0.7)
plt.title('Distribución de Meses de Nacimiento')
plt.xlabel('Mes')
plt.ylabel('Frecuencia')
plt.xticks(range(1, 13), ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun', 'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic'])
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Como podemos ver en el histograma exixsten datos que no tienen asignado un valor de 1 al 12 de acuerdo a su mes

```python
print(df['fn_mes'].unique())

print(df['fn_dia'].unique())

```

Existen valores que tienen un 99 en la columna de mes por lo que vamos a borrarlos, Tambien tienen el dia como 99, por lo que no nos sirven para este ejemplo
 y hacer de nuevo nuestro histograma

```python
meses_malos = (df['fn_mes'] == 99)

dias_malos = (df['fn_dia'] == 99)


df = df.drop(df[meses_malos].index)

df = df.drop(df[dias_malos].index)


# Crear un histograma de barras para la columna 'fn_mes'
plt.figure(figsize=(10, 6))
df['fn_mes'].value_counts().sort_index().plot(kind='bar', edgecolor='k', alpha=0.7)
plt.title('Distribución de Meses de Nacimiento')
plt.xlabel('Mes')
plt.ylabel('Frecuencia')
plt.xticks(range(0, 12), ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun', 'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic'])
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
